```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/drive/MyDrive/crypto/Download Data - CRYPTOCURRENCY_US_COINDESK_BTCUSD (1).csv")
df=df[1:]
```

```python
def settime(obj):
  a=list(obj)
  while "-" in a:
    a.remove("-")
  while "/" in a:
    a.remove("/")
  for i in range(len(a)):
    if a[i]==" ":
      a=a[0:i]
      break
  aa=""
  for i in a:
    aa+=i
  return aa

def setprice(obj):
  a=list(obj)
  while "," in a:
    a.remove(",")
  for i in range(len(a)):
    if a[i]==" ":
      a=a[0:i]
      break
  aa=""
  for i in a:
    aa+=i
  return int(aa)
```

```python
df["Date"]=df["Date"].apply(settime)
```

```python
df["Date"]=df["Date"].apply(lambda x: int(x))
```

```python
df["Open"]=df["Open"].apply(setprice)
df["Close"]=df["Close"].apply(setprice)
```

```python
y=df[["Close"]]
x=df.drop(columns=["High","Low","Close"])
df.drop(columns=["High","Low","Close"])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(xtrain)
xtrainS=scaler.transform(xtrain)
xtestS=scaler.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression(normalize=True)
model.fit(xtrainS,ytrain)
pred=model.predict(xtestS)
pred
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error,r2_score 
def evaluate(predd,ytestt):
  print("mean_absolute_error is ",mean_absolute_error(ytestt,predd))
  print("mean_squared_error is ",mean_squared_error(ytestt,predd))
  print("root_mean_squared_error is ",np.sqrt(mean_squared_error(ytestt,predd)))
  print("r2 score is",r2_score(ytestt,predd))
evaluate(ytest,pred)
```
